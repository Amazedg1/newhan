```python
import spotipy
import json
import webbrowser
import pandas as pd
```

```python
import urllib.request
import spotipy.util as util
```

```python
url="https://data.police.uk/api/crimes-at-location?date=2022-09&lat=52.629729&lng=-1.131592"
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
```

```python
crime_at_location = json.loads(response.read())
```

```python
number_of_crimes = len(crime_at_location)
```

```python
print(number_of_crimes)
```

```python
risk_factor = ""

if number_of_crimes >= 10:
    risk_factor = "Dangerous"
elif 5 <= number_of_crimes < 10:
    risk_factor = "High-risk"
elif 1 <= number_of_crimes < 5:
    risk_factor = "Caution"
else:
    risk_factor = "Safe"
```

```python
risk_factor
```

```python
credentials = "example_spotify.json"
with open(credentials, "r") as keys:
    api_tokens = json.load(keys)
```

```python
client_id = api_tokens["client_id"]
client_secret = api_tokens["client_secret"]
redirectURI = api_tokens["redirect"]
username = api_tokens["username"]
```

```python
api_tokens
```

```python
scope = 'user-read-private user-read-playback-state user-modify-playback-state playlist-modify-public'
token = util.prompt_for_user_token(username, scope, client_id=client_id,
                           client_secret=client_secret,
                           redirect_uri=redirectURI)
```

```python
sp = spotipy.Spotify(auth=token)
```

```python
devices = sp.devices()
print(devices)
deviceID = devices['devices'][0]['id']
```

```python
track_results = sp.search(q=risk_factor, type='track', limit=50)
```

```python

tracks = track_results['tracks']['items']
```

```python

tracks[0]
```

```python
# URI = uniform resource identifier
track_selection_list = []
track_art = []

for song in tracks:
    track_selection_list.append(song['uri'])
    track_art.append(song['album']['images'][0]['url'])
```

```python
my_playlist = sp.user_playlist_create(user=username, name=risk_factor, public=True,
                                      description="Crime song")
results = sp.user_playlist_add_tracks(username, my_playlist['id'], track_selection_list)
print(results)
```
